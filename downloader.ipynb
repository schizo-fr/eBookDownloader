```python
import urllib.request
urllib.request.urlretrieve("http://hydraprod.library.cornell.edu/fedora/get/seapage:A24_1/digitalImage", "00000001.jpg")
```

```python
import urllib.request
url = "http://hydraprod.library.cornell.edu/fedora/get/seapage:A24_%s/digitalImage"
dest = "jpg/page_{:0>2d}.jpg"
for i in range(1,81):    
    print(url % i + " ... " + dest.format(i)) 
    urllib.request.urlretrieve(url % i, dest.format(i)) 
    print(dest.format(i) + " OK")
```
